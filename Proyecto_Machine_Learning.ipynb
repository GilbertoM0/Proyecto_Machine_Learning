# Proyecto de Machine Learning

------------------------------------------

## Introducción a mi proyecto

En los procesos de selección y evaluación personal, las expresiones faciales juegan un papel clave para interpretar las emociones de los candidatos. Sin embargo, esta interpretación suele ser subjetiva y depende de la experiencia del entrevistador. Para abordar este problema, propongo desarrollar una aplicación llamada FaceInterview que utilice visión por computadora e inteligencia artificial para analizar las emociones de los candidatos durante una entrevista. El modelo permitirá identificar emociones como felicidad, tristeza, enojo o sorpresa en tiempo real, proporcionando retroalimentación precisa y objetiva para mejorar la toma de decisiones en los procesos de selección.

------------------------------------------

### Objetivo General
Desarrollar una aplicación llamada FaceInterview que utilice visión por computadora e inteligencia artificial para analizar en tiempo real las expresiones faciales de los candidatos durante una entrevista, identificando las emociones predominantes para mejorar la evaluación y toma de decisiones en los procesos de selección.

### Objetivos Específicos
- Desarrollar un modelo de reconocimiento de emociones mediante el uso de redes neuronales convolucionales (CNN) y bibliotecas como TensorFlow y DeepFace.
- Recolectar y preprocesar datos utilizando conjuntos de datos públicos (FER-2013, CK+ o AffectNet) para entrenar y mejorar el modelo.
- Implementar un sistema de análisis en tiempo real que capture imágenes desde la cámara o archivos y analice las emociones de manera rápida y precisa (<500ms).
- Crear una interfaz amigable que muestre las emociones detectadas y permita obtener reportes o retroalimentación sobre las respuestas emocionales del candidato.
- Evaluar el rendimiento del modelo utilizando métricas como precisión (accuracy), F1-Score y matriz de confusión para asegurar un rendimiento superior al 85%.


------------------------------------------

1. ¿Qué problema se quiere resolver?
   
El problema que queremos resolver es la evaluación automática de emociones durante entrevistas o interacciones mediante el análisis de expresiones faciales. El objetivo es ayudar a identificar el estado emocional de una persona (como nerviosismo, confianza o interés) para mejorar la toma de decisiones en procesos de selección o evaluación personal.

------------------------------------------

2. ¿Tienes datos suficientes y de calidad?
Puedo empezar utilizando conjuntos de datos públicos que investigue como:

**- FER-2013 (Facial Emotion Recognition) – 35,887 imágenes etiquetadas con 7 emociones.**


**- CK+ (Extended Cohn-Kanade) – Datos para expresión facial dinámica.**

**- AffectNet – Más de 1 millón de imágenes con anotaciones emocionales.**

Sin embargo, si quiero un modelo más preciso y ajustado a entrevistas reales, podría:

- Ampliar el conjunto de datos capturando imágenes reales en situaciones de entrevista.
- Realizar aumentación de datos para cubrir diferentes ángulos, iluminación y posiciones faciales.


------------------------------------------

3. ¿Qué métrica indicará el éxito del modelo?
Las métricas que indicarán el éxito del modelo incluyen:

- Accuracy (Precisión): Porcentaje de emociones detectadas correctamente.
- F1-Score: Para medir el equilibrio entre precisión y exhaustividad.
- Confusion Matrix: Para ver qué emociones se confunden entre sí.
- Tiempo de respuesta: Para asegurar que el análisis sea en tiempo real (idealmente <500ms por análisis).
- Éxito → Si el modelo tiene una precisión superior al 85% y el análisis es rápido (<500ms), podemos considerarlo exitoso para una aplicación en entrevistas.
